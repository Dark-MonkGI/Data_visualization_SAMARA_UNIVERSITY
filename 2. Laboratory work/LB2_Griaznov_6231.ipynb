### Что делает geom_col()? Чем он отличается от geom_bar()?

#### Type your code here...
Есть два типа гистограмм: geom_bar() и geom_col() . geom_bar() делает высоту столбца пропорциональной количеству случаев в каждой группе (или, если задана эстетика весов, сумме весов). Если вы хотите, чтобы высота столбцов представляла значения в данных, используйте вместо этого функцию geom_col() .

geom_bar( ) по умолчанию использует stat_count() : он подсчитывает количество наблюдений в каждой позиции x.

по сути, geom_col(…) равно geom_bar(stat = «identity») 

```python
ggplot(data=diamonds) +\
geom_col(mapping=aes(x="cut", y="table"))
```

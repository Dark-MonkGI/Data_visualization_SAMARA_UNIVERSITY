```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/employee.csv')
data = data.loc[:199,:]
print(data.shape)
```

```python
data['Department'].value_counts().plot(kind = 'bar', color = 'lightblue')
plt.xlabel('Отделы')
plt.xticks(rotation = 360)
plt.title('Количество сотрудников в каждом отделе', fontsize =12)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (8, 4)
data['EducationField'].value_counts().plot(kind = 'bar', color = 'orange')
plt.xlabel('Различные области образования')
plt.xticks(rotation = 360)
plt.title('Количество сотрудников из разных областей образования', fontsize =12)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 4)
x = data.pivot_table(index = ['Department','EducationField'], values = ['MonthlyRate'], aggfunc = 'mean')
x.plot(kind = 'bar', stacked = False)
plt.xticks(rotation = 90)
plt.show()
```

```python
x = pd.crosstab(index = data['Department'],
                columns = data['EducationField'],
                values = data['MonthlyRate'],
                aggfunc = 'mean')
x
```

```python
plt.rcParams['figure.figsize'] = (8, 4)
x.plot(kind = 'bar', stacked = False)
plt.xticks(rotation = 20)
plt.show()
```

### Scatter Plot

```python
data = pd.read_csv('/employee.csv')
```

```python
plt.scatter( data['TotalWorkingYears'],
            data['MonthlyRate'],
            color = 'red')

plt.ylabel('Месчяный доход')
plt.xlabel('Выработка лет')
plt.title('Отношение дохода и опыта')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (16, 6)

plt.scatter( data['TotalWorkingYears'],
            data['MonthlyRate'],
            s = data['DistanceFromHome'],
            color = 'green')
plt.ylabel('Месячный доход')
plt.xlabel('Время работы')
plt.title('Отношение дохода к выслуге', fontsize = 20)
plt.show()
```
